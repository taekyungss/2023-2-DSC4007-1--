```python
from konlpy.corpus import kolaw
const_doc = '''
어려움 있어
내가 일을 잃었어 실직한지 좀 됐어
어 괜찮은데 지금 실직한지 좀 오래 돼 가지고 일을 제대로 못 구하고 있어 그래서 공과금도 못 내고 있어
그것뿐만 아니라 밥 먹고 살아야 되는데 밥 먹기도 좀 힘든 거 같아 그 정도로 돈이 없어
맞아 근데 지금 혼자 할머니랑 엄마랑 같이 살고 있어
아 그렇구나 혹시 더 알려 줄 수 있나
'''

print(type(const_doc))
print(len(const_doc))
print(const_doc[:10000])
```

```python
t.pos(const_doc)
```

```python
a = [word for word, pos in t.pos(const_doc) if pos in ('Adjective')]
a
```

```python
tokens_const = [word for word, pos in t.pos(const_doc) if pos not in ('Josa','Verb','Eomi','Foreign')]

tokens_const
```

```python
# tokens_const = [token for token in tokens_const if len(token)  > 1] #한 단어인 명사는 제외 (애, 것, 함, 제, 장)
tokens_const = [token for token in tokens_const if token not in ['선생님','좀','혹시','지금','이제','상담사','조금','말씀','저희','계층','사회','강민지','못','그','것','있어']]
print('#토큰의 수:', len(tokens_const))
print('#앞 100개의 토큰')
print(tokens_const[:100])
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

const_cnt = {}
for word in tokens_const:
    const_cnt[word] = const_cnt.get(word,0) + 1
    
def word_graph(cnt, max_words=10):
    sorted_w = sorted(cnt.items(), key=lambda kv: kv[1])
    print(sorted_w[-max_words:])
    n, w = zip(*sorted_w[-max_words:])
    
    plt.barh(range(len(n)), w,tick_label=n)
    plt.show()
    
word_graph(const_cnt, max_words=20)
```

```python
import seaborn as sns
import pandas as pd
data = pd.DataFrame({
    'Category': ['타 복지 사업', '고용 위기', '생활 수준 추정','상황 변동','요금 체납','기타 민감 정보'],
    'Weighted Count': [count_list_1 * weight_list_1, count_list_2 * weight_list_2, count_list_3 * weight_list_3,
                       count_list_4*weight_list_4, count_list_5*weight_list_5, count_list_6*weight_list_6]
})
data['Rank'] = data['Weighted Count'].rank(ascending=False, method='dense').astype(int)
palette = sns.color_palette("viridis", n_colors=len(data))


# 그래프 그리기
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Weighted Count', data=data, palette=palette)
plt.title('Weighted Count for Each Category')
plt.show()
```
